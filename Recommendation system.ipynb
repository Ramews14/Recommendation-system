```python
# Create a user-item interaction matrix
user_movie_matrix = df_cleaned.pivot_table(index='userId', columns='title', values='rating')

# Preview the user-item matrix
print(user_movie_matrix.head())
```

```python
from scipy.sparse import csr_matrix
from sklearn.metrics.pairwise import cosine_similarity

# Convert the user-movie matrix to a sparse matrix
user_movie_sparse_matrix = csr_matrix(user_movie_matrix.fillna(0))

# Compute cosine similarity instead of Pearson to save memory
user_similarity_sparse = cosine_similarity(user_movie_sparse_matrix)

# Convert back to a DataFrame for ease of use
user_similarity_sparse_df = pd.DataFrame(user_similarity_sparse, index=user_movie_matrix.index, columns=user_movie_matrix.index)

# Preview the sparse similarity matrix
print(user_similarity_sparse_df.head())

```

```python
def get_sparse_recommendations(user_id, user_movie_matrix, user_similarity_sparse_df, n_recommendations=5):
    # Select the user’s ratings from the user_movie_matrix
    user_ratings = user_movie_matrix.loc[user_id].dropna()
    print("User Ratings:")
    print(user_ratings)

    # Find similar users from the sparse similarity matrix
    similar_users = user_similarity_sparse_df[user_id].sort_values(ascending=False)
    print("\nSimilar Users:")
    print(similar_users)

    # Remove the user themselves
    similar_users = similar_users.drop(user_id, errors='ignore')

    # Weighted average of ratings by similar users
    movie_scores = pd.Series(dtype=float)

    for similar_user, similarity_score in similar_users.iteritems():
        similar_user_ratings = user_movie_matrix.loc[similar_user].dropna()
        print(f"\nRatings of Similar User {similar_user}:")
        print(similar_user_ratings)
        for movie, rating in similar_user_ratings.iteritems():
            if movie not in user_ratings:
                if movie not in movie_scores:
                    movie_scores[movie] = similarity_score * rating
                else:
                    movie_scores[movie] += similarity_score * rating

    print("\nMovie Scores:")
    print(movie_scores)

    # Sort movie scores and return top N recommendations
    recommendations = movie_scores.sort_values(ascending=False).head(n_recommendations)
    return recommendations

# Example: Get top 5 movie recommendations for a specific user (userId = 1)
recommendations = get_sparse_recommendations(1, user_movie_matrix, user_similarity_sparse_df, n_recommendations=5)
print("\nTop Recommendations:")
print(recommendations)
```
